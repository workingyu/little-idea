```python
from gtts import gTTS
import IPython.display
```

```python
message = "你说话好机车啊 很讨厌"
tts = gTTS(message, lang='zh-CN')
tts.save('gTTS_test.mp3')
IPython.display.Audio('gTTS_test.mp3')
```
